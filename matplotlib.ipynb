```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
x = np.arange(1, 25)
y = np.cos(2*np.pi*x)*np.exp(-x)
```

```python
# Line Plot
plt.figure(figsize=(10,5))
plt.plot(x, y, color="darkorange", linewidth=2, marker='^')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("x vs cos(x)")
# plt.yticks([-1, 0, 1])
# plt.xticks([0,2,4])
plt.grid(True)
plt.show()
```

```python
# Line Plot
plt.figure(figsize=(10,5))
plt.plot(x, np.cos(x), color="darkorange", linewidth=2, marker='^', label="cosX")
plt.plot(x, np.sin(x), color="lightblue", linewidth=2, marker='^', label="sinX")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("x vs cos(x)")
plt.legend()
plt.grid(True)
plt.xlim([5,20])
plt.ylim([-0.75,0.75])
plt.show()
```

```python
# Bar 
age = [10,12,14,16,18,20]
students = [30,50,77,22,33,89]
plt.bar(age, students)
plt.title("Age Distribution")
plt.show()
```

```python
# Bar 
lang = ['Python', 'Java', 'C++']
demands = [150,50,77]
plt.barh(lang, demands)
plt.title("Age Distribution")
plt.show()
```

```python
# Histogram
plt.hist(age, bins=5)
plt.show()
```

```python
# Scatter
y = np.random.random(24)
plt.scatter(x, y, color="purple", marker="*")
plt.show()
```

```python
# Pie
sports = ["Cricket", "Football", "Poetry", "VolleyBall"]
participation = [84, 74, 23, 50]
plt.pie(participation, labels=sports, explode=(0,0,0,0.1), shadow=(0,0,0,0.5),autopct='%1.3f%%')
```
